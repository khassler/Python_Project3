# Scientific Question: Did the ALX1 gene that encodes a trancription factor affecting the craniofacial development in the independent species of Finches show genomic evidence of adaptive radiation through genetic differences in the ALX1 gene of each species?

The ALX1 gene is encompassed by a 240 kilobase haplotpye and codes for a transcription factor that is crucial for nomral craniofacial development in vertebrates.  Mild mutations in this gene are what contribute to the facial diversity of finches. The ALX1 protein is a transcritpion factor that binds to the DNA an can control the activity of certain genes, expecially the genes that regulate cell growth and proliferation (Lamichhaney et al., 2015)  

There are many researchers that have studied the the ALX1 gene and its effects on the craniofacial development of Finches and many other species.  COME BACK TO THIS

# Hypothesis: If the ALX1 gene is correlated with adaptive radiation in Finches, then a change in the ALX1 protein structure and genomic makeup will be different between each species of Finch.

There are 458 different ALX1 genes that have been sequenced from 458 different species.  Only 3 of those ALX1 genes that have been identified were the finch species.  We were able to determine the gene sequences through the GenBank function on NCBI.  I then saved the Gene sequences as an SDS final to use for multiple sequence alignment, the first bioinformatics analysis done in this notebook. 

RNA sequence alignments generally align three or more biological sequences that have a similar length.  It can be used to compare and contrast the three biological sequences of each ALX1 gene.

The contrast between the ALX1 gene of each species of finch can be better understood through the use of a Pariwise Sequence Alignment.  The second bioinformatics method done in the notebook.

```python
def createMatrix(seq_x, seq_y):
    seq_x_list = list(seq_x)
    seq_y_list = list(seq_y)
    matrix = np.zeros((len(seq_x_list), len(seq_y_list)))
    for i in range(len(seq_x_list)):
        for j in range(len(seq_y_list)):
            if seq_x_list[i] == seq_y_list[j]:
                matrix[i][j] = 1
    return matrix
matrix = createMatrix(Geospiza_fortis_seq, Taeniopygia_seq)
ax = sns.heatmap(matrix,
                 xticklabels=False,
                 yticklabels=False,
                 cbar=False,
                 center=0)
```

# Experimenting with a dot plot since my Heatmap is not appearing

```python
def delta(x,y):
    return 0 if x == y else 1
def M(seq1,seq2,i,j,k):
    return sum(delta(x,y) for x,y in zip(seq1[i:i+k],seq2[j:j+k]))
def makeMatrix(seq1,seq2,k):
    n = len(seq1)
    m = len(seq2)
    return [[M(seq1,seq2,i,j,k) for j in xrange(m-k+1)] for i in xrange(n-k+1)]
def plotMatrix(M,t, seq1, seq2, nonblank = unichr(0x25A0), blank = ' '):
    print(' |' + seq2)
    print('-'*(2 + len(seq2)))
    for label,row in zip(seq1,M):
        line = ''.join(nonblank if s < t else blank for s in row)
        print(label + '|' + line)
def dotplot(seq1,seq2,k = 1,t = 1):
    M = makeMatrix(seq1,seq2,k)
    plotMatrix(M, t, seq1,seq2)
```
